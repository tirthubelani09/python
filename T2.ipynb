```python
import pandas as pd
df = pd.read_csv('windpower.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
x = df[['Wind speed (m/s)']]
y = df['Power (kW)']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
predictions =[]
predictions.append(y_test)
mse=[]
for i in range(3,7):
    poly = PolynomialFeatures(degree=i)
    x_train_poly = poly.fit_transform(x_train)
    x_test_poly =poly.fit_transform(x_test)
    model = LinearRegression()
    model.fit(x_train_poly,y_train)
    y_pred = model.predict(x_test_poly)
    predictions.append(y_pred)
    print(f'The Coefficients for degree {i} are ',model.coef_)
    print(f'the MSE for degree {i} is ',mean_squared_error(y_test,y_pred))
    print(f'The R2 score for degree {i} is',r2_score(y_test,y_pred))
    mse.append(mean_squared_error(y_test,y_pred))
    x_new = pd.DataFrame({'Wind Speed(m/s)':[5,10,15,20]})
    x_new_poly = poly.fit_transform(x_new)
    y_new = model.predict(x_new_poly)
    print(f'Predictions for windpower 5,10,15,20 for degree {i} are',y_new)
    print('-----------------------------------------------------------------')
```

```python
import matplotlib.pyplot as plt
for i in predictions:
    plt.scatter(x_test,i)
plt.show
```

```python
deg = [3,4,5,6]
plt.plot(deg,mse)
plt.show()
print('Degree 6 has lowest mean sqaured error')
```
